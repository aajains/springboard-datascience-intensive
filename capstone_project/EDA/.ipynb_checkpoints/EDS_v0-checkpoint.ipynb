### Addressing columns with missing values

#### CRSElapsedTime: 
This is the flight elapsed time generated from the computerized reservation system (CRS). This elapsed time also includes the time-zone difference. When do we get null values for this column?

```python
df[pd.isnull(df.CRSElapsedTime)] 
```

```python
sum(pd.isnull(df.CRSElapsedTime))
```

It is only Spirit Airlines (NK) for which we see missing values for CRSElapsedTime (total 12 times: 6 times in 2015 and 6 times in 2016). Note the airline name using the Carrier code (like NK) can be found [here](http://www.iata.org/publications/Pages/code-search.aspx). More importantly, all these NK flights were cancelled, except one from FLL to LGA on 2015-04-20. Does this mean that whenever the Spirit Airlines bookings do not tell the CRSElapsedTime, their flights get canceled? This could be a possibility and a good indicator if we are only interested in predicting cancellation likelihood for Spirit Airlines. But we are interested in all airlines. Moreover, the CRSElapsedTime can be highly correlated with "Distance". The higher the flight distance, the higher the CRSElapsedTime will be. If we do have such correlation,  then we can 

(i) keep the "Distance" 

(ii) remove the "CRSElapsedTime" column

(iii) create a new colum "CRSElapsedTimeNaN": this will have 1 if the data is missing and 0 otherwise. 

But before doing that, can we confirm if we do have such a correlation?  

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.plot(df.CRSElapsedTime, df.Distance, marker='o', linestyle='', ms=8, alpha=0.4, color=tableau20[18])
plt.xlabel("CRSElapsedTime")
plt.ylabel("Distance")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

Visually, we do observe a positive correlation with few outliers. There are some distances for which there are many CRSElapsedTime. This could be due to time-zone difference, or due to changes in reservation system for same distance flights. Let us quickly perform a hypothesis test to see if there is any correlation. We can calculate a Pearson correlation coefficient. However, this calculation requires that both CRSElapsedTime and Distance are normally distributed. So, let us see their distribution first.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,8))
sns.distplot(df.CRSElapsedTime[pd.notnull(df.CRSElapsedTime)], 500, ax=ax1, color=tableau20[6])
ax1.set_title("CRSElapsedTime Distribution")
sns.distplot(df.Distance, 500, ax=ax2, color=tableau20[8])
ax2.set_title("Distance Distribution")
plt.show()
```

These do not look normal distribution. Lets test them:

```python
from scipy import stats
```

```python
stats.normaltest(df.CRSElapsedTime[pd.notnull(df.CRSElapsedTime)])
```

```python
stats.normaltest(df.Distance)
```

Clearly, they are not normally distributed. So, we cannot use the Peasrson's correlation coefficient. We could rather use a non-parametric test such as Spearman's test. Lets calculate the Spearman's correlation coefficient and associated p-value.

```python
stats.spearmanr(df.CRSElapsedTime[pd.notnull(df.CRSElapsedTime)], df.Distance[pd.notnull(df.CRSElapsedTime)])
```

We can reject thhe null that there is no correlation, and accept the alternate hypothesis that ***there is indeed very strong position correlation (0.98) between the two variables***. Hence, we can drop CRSElapsedTime from our data set (***<font color='red' size=4>which we will do later</font>***) and add a new column "CRSElapsedTimeNaN" with 0s and 1s.

```python
df["CRSElapsedTimeNaN"] = 0
df.CRSElapsedTimeNaN[pd.isnull(df.CRSElapsedTime)] = 1
```

#### TailNum:

Tail number is just like the registration number on our cars. This number is specific to a plane. The same plane can be used for different flights. For example, the plane with TailNum = N787AA in 2015 has been used in two different flights (different routes) such as JFK-LAX, JFK-SFO:

```python
df[(df.TailNum == "N787AA") & (df.Year == 2015)][["TailNum", "FlightDate", "Origin", "Dest"]].head(15)
```

This means that the tail number is not consistent for a given route. So, we can drop TailNum too (***<font color='red' size=4>which we will do later</font>***).

These 5 columns have exactly same number of missing values in them, and they are all related with actual departure time and departure delays (in minutes).
```python
all(df.Cancelled[pd.isnull(df.DepartureDelayGroups)] == 1)
```

If we think about our problem (in broad sense) carefully, we want to predict if the flight will be canceled before 2-3 days. At the time of making this prediction, it is impossible to know actual departure time and departure delays. So, there is no point in keeping these 5 columns for the modeling purpose. However, we can use them to engineer some interesting features, which we will discuss later. So, for the time being lets kepp them.

#### TaxiOut, WheelsOff:


These two columns have exactly same number of missing. TaxiOut is the time (in minutes) a flight spends after the pilot releases plane's parking brake from the gate to the time plane takes off (leaves ground). The time when the plane leaves the ground is called WheelsOff time. The null fractions for these two columns are smaller than overall cancellation rate, so we guess the same as above and confirm them below:

```python
all(df.Cancelled[pd.isnull(df.TaxiOut)] == 1)
```

```python
all(df.Cancelled[pd.isnull(df.WheelsOff)] == 1)
```

We have the same reasoning as above to not remove these columns.

#### WheelsOn, TaxiIn, ArrTime:


The WheelsOn is the time when the plane touches down the ground. TaxiIn the duration (in minutes) between WheelsOn and the time when pilot applies parking brake at the gate. ArrTime is the actual arrival time. The null fractions for these columns are bit higher than cancellation rate. This means that there are flights that were not canceled but even then we got missing values for these columns. What are those cases? 

```python
len(df[(pd.isnull(df.WheelsOn)) & (df.Cancelled == 0)])
```

```python
len(df[(pd.isnull(df.TaxiIn)) & (df.Cancelled == 0)])
```

```python
len(df[(pd.isnull(df.ArrTime)) & (df.Cancelled == 0)])
```

#### ArrDelay, ArrDelayMinutes, ArrDel15, ArrivalDelayGroups, ActualElapsedTime, AirTime:

First 4 columns pertain to arrival time and arrival delay (in minutes). ActualElapsedTime is self explainatory and the AirTime is literally the time a flight spends in the air (in minutes). Again, all these columns are not directly useful in predicting cancellation likelihood (as we discussed before). However, they might be useful in feature engineering later, so we will kepp them too.

For comparing cancellation rates based on different categories, we create a generic plot function below first:

```python
def plot_cancel_count_by_temporal(by, xlab, ylab, title, ylim):
    plt.subplots(1, 1, figsize=(10, 4))

    ax = plt.subplot(1, 1, 1)
    (100*df.groupby(by)["Cancelled"].sum() / df.groupby(by)["Cancelled"].count()).plot(kind="bar", 
                                                                                         color=tableau20[16])
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.ylim(ylim)
    plt.title(title)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
```

```python
plot_cancel_count_by_temporal("Year", "Year", "% Cancellation", "% Cancellation by year", [0, 1.85])
```

#Since the number of unique cities are 322, we will end up with quite complicated model if we add this variable in the model. To simply, we might coarse-grain the airport location information by considering the OriginState variable.

```python
def plot_cancel_count_by_spatial(by, xlab, ylab, title, ylim, sort):
    fig, ax = plt.subplots(1, 1, figsize=(20, 12))
    can_rate = (100*dfs.groupby(by)["Cancelled"].sum() / dfs.groupby(by)["Cancelled"].count())
    can_rate_sorted = can_rate.sort_values(ascending=False)
    if sort:
        ax = can_rate_sorted.plot(kind='bar',facecolor=tableau20[16])
    else:
        ax = can_rate.plot(kind='bar',facecolor=tableau20[16])
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.ylim(ylim)
    plt.title(title)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    plt.setp( ax.xaxis.get_majorticklabels(), rotation=90)
    #ax.invert_yaxis()
```

We now use PCA to better understand which predictor variable explains more variance. However, before doing that we first split our data into train and test sets, and then normalize all our predictor variables.

***Splitting***

```python
X_train, X_test, y_train, y_test = train_test_split(dfsr.drop(["Cancelled"], axis=1), 
                                                    dfsr.Cancelled, test_size=0.5, random_state=10)
```

***Scaling***

```python
scaler = MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

***PCA***

```python
pca = PCA().fit(X_train_scale)
```

***Scree plot***

```python
plt.plot(pca.explained_variance_, c=tableau20[6])
plt.xlabel("Number of components")
plt.ylabel("Exaplained variance")
```

```python
plt.plot(pca.explained_variance_ratio_.cumsum(), c=tableau20[6])
plt.xlabel("Number of components")
plt.ylabel("Cumulative explained variance")
```

There is no distinct boundary between high and low values of explained variance, except last few components. We could go with 60-80 components to simplify the model little bit and loose the interpretability of variables. Let us not do that at this stage, and stick with all components, i.e. let us proceed without PCA components (or with original data set). Also, even if wanted to visualize first few components we cannot do that because to have even 50% of the variance we would need about 50 components! 2-3 first components represent only about 20% of the total varience.
